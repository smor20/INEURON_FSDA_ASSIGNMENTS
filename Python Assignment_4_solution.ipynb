### 1. Write a Python Program to Find the Factorial of a Number?

```python
def factorial(num):
    if (num < 1):
        return 1
    else:
        return num*factorial(num-1)
num = int(input('Enter a number: '))
value = factorial(num)
print(f'The Factorial of {num} is {value}')
```

### 2. Write a Python Program to Display the multiplication Table?

```python
def get_table(base,val):
    for x in range(1,val+1):
        print(f'{base} X {x} = {base*x}')

num = int(input('Enter a number: '))
values = int(input('Enter no of val: '))
get_table(num,values)
```

### 3. Write a Python Program to Print the Fibonacci sequence?


```python

nterms = int(input("How many term: "))


n1, n2 = 0, 1
count = 0


if nterms <= 0:
   print("Please enter a positive integer")

elif nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)

else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1)
       nth = n1 + n2
       
       n1 = n2
       n2 = nth
       count += 1
```

### 4. Write a Python Program to Check Armstrong Number?

```python
def find_as_num():
    in_num = input('Enter a number: ')
    sum = 0
    for char in range(len(in_num)):
        sum = sum + pow(int(in_num[char]),3)
    if sum == int(in_num):
        print(f'{in_num} is a Armstrong Number')
    else:
        print(f'{in_num} is a Not Armstrong Number')


find_as_num()
```

### 5. Write a Python Program to Find Armstrong Number in an Interval?

```python
def as_nums(in_num, storage):
    sum = 0
    for char in range(len(in_num)):
        sum = sum + pow(int(in_num[char]),3)
    if sum == int(in_num):
        storage.append(int(in_num))

start = int(input('Enter the Starting number: '))
end = int(input('Enter the Ending number: '))
list_of_armstrong = []

if start > end:
    print("Start Interval Cannot be Greater than End Interval")
else:
    for number in range(start,end+1):
        as_nums(str(number),list_of_armstrong)
    print(f'The Armstrong numbers between {start} and {end} are {list_of_armstrong}')
```

### 6. Write a Python Program to Find the Sum of Natural Numbers?

```python
def sum_n_numbers(num):
    sum = num*((num+1)/2)
    print(f'Sum of {num} natural numbers is {sum}')
    
num = int(input('Enter a number: '))
sum_n_numbers(num)
```
